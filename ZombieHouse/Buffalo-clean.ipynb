```python
import sodapy
from sodapy import Socrata
import os
import geopandas as gpd
import pandas as pd
import numpy as np
```

### Load Code violation data set

```python
socrata_domain='data.buffalony.gov'
socrata_dataset_identifier='ivrf-k9vm'
app_token=os.environ.get('SODAPY_APPTOKEN')
client=Socrata(socrata_domain,app_token,timeout=30)
dict=client.get(socrata_dataset_identifier,content_type='geojson?$limit=1000000',\
select="LOCATION,ADDRESS,CODE,CODE_SECTION")
gdf_code=gpd.GeoDataFrame.from_features(dict)
gdf_code.tail()
```

### Load buffalo-clean data set

```python
buffalo_clean=pd.read_csv('buffalo-clean.csv',encoding = "utf-8")
```

```python
buffalo_clean_good=buffalo_clean[['address','vacant_','latitude_','longitude_']] ## Select the appropriate data column
```

```python
buffalo_clean_good['address'].value_counts()
```

### Data cleaning

```python
buffalo_clean_good_1=buffalo_clean_good.dropna(subset=['latitude_','longitude_'])   
buffalo_clean_good_2=buffalo_clean_good_1.drop_duplicates(['address']) #Delete duplicate values
buffalo_clean_good_2=buffalo_clean_good_2.rename(columns={'address':'ADDRESS'}) 
```

```python
buffalo_clean_good_2
```

```python
df1=pd.merge(buffalo_clean_good_2,gdf_code,how='left',on='ADDRESS') #Merge code violation data set and buffalo-clean data set
```

```python
df1 #Most of the addresses in Table Buffalo-clean do not have Code Violation Information
```

```python
address_code=df1[df1['CODE'].notnull()] #Find  houses that has Code violation
```

```python
address_code # Just 2980 addresses have Code Violation info 
```

```python
address_code_group=address_code.groupby(['ADDRESS']).aggregate(lambda x:','.join(x)) 
# Gourp by the table by ADDRESS and aggregate Code Violation info on each address
```

```python
address_code_group=address_code_group.reset_index()
```

```python
address_code_group 
```

```python
#Add some additional information to the 'address_code_group' table
address_code_all=pd.merge(address_code_group,address_code,how='left',on='ADDRESS')
address_code_all=address_code_all.drop_duplicates(['ADDRESS'])
address_code_all=address_code_all.drop(['CODE_y','CODE_SECTION_y'],axis = 1)
```

```python
address_code_all #This table is part of the Buffalo-clean data set and  contains the corresponding Code Violation information
```

### Now split the CODE_x column

```python
def count_type(sqe):
    c={}
    for i in sqe.split(','):
        if i not in c:
            c[i]=1
        else:
            c[i]=c[i]+1
    return c
```

```python
address_code_all['CODE']=address_code_all['CODE_x'].map(count_type)
```

```python
address_code_all = pd.concat([address_code_all, address_code_all['CODE'].apply(pd.Series)], axis = 1).drop('CODE', axis = 1)
```

```python
address_code_all #The columns of the table represent the Code Cheesecake types, and their values represent quantity
```

```python
address_code_all_nog=address_code_all.drop(['geometry'],axis=1) #In order for the next step to add 0 instead of null
```

```python
address_code_all_nog=address_code_all_nog.fillna(0)
```

```python
address_code_all_nog
```

```python
training_set=address_code_all_nog[address_code_all_nog['vacant_']!=0] #Filter the training set 
#select the data with known vacant data and non-vacant data
```

```python
training_set
```

## Model ( Decision Tree Algorithm )

```python
from sklearn.tree import DecisionTreeClassifier as DTC, export_graphviz
from sklearn import tree
from sklearn.model_selection import train_test_split
import graphviz
```

```python
df_X=training_set.iloc[:,6:]
df_X[df_X>0]=1
```

```python
df_Y=training_set.iloc[:,3]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(df_X,df_Y,test_size=0.3)
```

```python
dtc=DTC(criterion='entropy')
dtc.fit(X_train,Y_train)
print('Accuracy:',dtc.score(X_test,Y_test))
with open('tree2.dot','w') as f:
    f=export_graphviz(dtc,feature_names=df_X.columns,out_file=f)
```
